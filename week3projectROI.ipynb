```python
class ReturnOnInvestment():
    def __init__(self, value):
        self.value = {}
        
    def totalInvestment(self):
        home_value = int(input("Enter home value: "))   
        down_payment = int(input("Enter down payment: "))
        closing_cost = int(input("Enter closing cost: "))
        rehab_budget = int(input("Enter rehab budget(any money spent on preparing house to be rentable/sellable): "))
        misc_investment = int(input("Enter any miscellaneous investment costs: "))
        total_investment = down_payment + closing_cost + rehab_budget + misc_investment
        self.value['Home Value'] = home_value
        self.value['Down Payment'] = down_payment
        self.value['Closing Cost'] = closing_cost
        self.value['Rehab Budget'] = rehab_budget
        self.value['Misc. Investment Cost'] = misc_investment
        self.value['Total Investment'] = total_investment
    
    def totalIncome(self):
        print("Please enter values based on per-month projections:")
        rent_income = int(input('Enter rent income: '))
        misc_income = int(input('Enter miscellaneous income: '))
        self.value['Annual Rent Income'] = rent_income * 12
        self.value['Annual Misc. Income'] = misc_income * 12
        self.value['Total Annual Income'] = (rent_income * 12) + (misc_income * 12)
          
    def totalExpenses(self):
        print("Please enter values based on annual projections:")
        
        capExp = int(input("Enter estimated capital expenditure (any money used to add/improve property beyond commond repair/maintenance): "))
        
        while True:
            response = input("Would you like to manually input property insurance expense? If no, type 'no' and expense will be automatically calculated based on the Tennessee average. If yes, type 'yes'. ")
            if response.lower() == 'no':
                global property_ins
                property_ins = self.value['Home Value']*0.0065
                break
            elif response.lower() == 'yes':
                property_ins = int(input("Enter property insurance:"))
                break
            else: 
                print("Please enter 'yes' or 'no'.")
        
        while True:
            response = input("Would you like to manually input property tax expense? If no, type 'no' and expense will be automatically calculated based on the Tennessee average. If yes, type 'yes'. ")
            if response.lower() == 'no':
                global property_tax
                property_tax = self.value['Home Value']*0.0064
                break
            elif response.lower() == 'yes':
                property_tax = int(input("Enter property tax:"))
                break
            else: 
                print("Please enter 'yes' or 'no'.")             
        
        while True:
            response = input("Would you like to manually input vacancy expense? If no, type 'no' and expense will be automatically calculated as 5% of monthly rental income. If yes, type 'yes'. ")
            if response.lower() == 'no':
                global vacancy
                vacancy = self.value['Annual Rent Income']*0.005
                break
            elif response.lower() == 'yes':
                vacancy = int(input("Enter vacancy expense:"))
                break
            else:
                print("Please enter 'yes' or 'no'.")

        print("\nPlease enter values based on per-month projections: ")
        hoa = int(input("Enter HOA cost: "))
        monthly_repairs = int(input("Enter estimated repairs/maintence cost: "))
        property_mgmt = int(input("Enter cost of hiring a property manager: "))                          
        mortgage = int(input("Enter mortgage payment: "))
        misc_exp = int(input("Enter any miscellaneous expenses: "))

        self.value['Annual Capital Expenditure'] = capExp
        self.value['Annual Property Insurance'] = property_ins
        self.value['Annual Property Tax'] = property_tax
        self.value['Annual Vacancy Expense'] = vacancy
        self.value['Annual HOA Expense'] = hoa * 12
        self.value['Annual Repairs'] = monthly_repairs * 12
        self.value['Annual Property MGMT Expense'] = property_mgmt * 12
        self.value['Annual Mortgage'] = mortgage * 12
        self.value['Annual Misc. Expense'] = misc_exp * 12
        self.value['Total Annual Expenses'] = property_ins + property_tax + vacancy + (hoa*12) + (monthly_repairs*12) + (property_mgmt*12) + (mortgage*12) + (misc_exp*12)
                                
    def returnCalculated(self):
        profitabilityCalc = (self.value['Total Annual Income'] - self.value['Total Annual Expenses'])/(self.value['Total Investment'])*100
        cashFlow = (self.value['Total Annual Income'] - self.value['Total Annual Expenses'])
        print("\nThese are your inputs and total values of: investment, income, and expenses\n")
        for key in self.value:    
            print(f"{key} : {self.value[key]}")
        
        print("\nFinal Report:")
        for key in {self.value['Total Investment']}:
            print(f"Total Investment : {self.value['Total Investment']}")
        for key in {self.value['Total Annual Income']}:
            print(f"Total Annual Income : {self.value['Total Annual Income']}")
        for key in {self.value['Total Annual Expenses']}:
            print(f"Total Annual Expenses : {self.value['Total Annual Expenses']}")
        
        print(f"Annual Cash Flow: {cashFlow}")
        print(f"ROI for this property is: {str(round(profitabilityCalc, 2))}%")


        
        
# runs the program           
def findReturnOnInvestment():
    roi = ReturnOnInvestment({})
    
    print("Welcome to the Return On Investment (ROI) calculator. Please follow these steps to calculate the ROI on any property!")
    print("\n")
    roi.totalInvestment()
    print("\n")
    roi.totalIncome()
    print("\n")
    roi.totalExpenses()
    print("\n")
    roi.returnCalculated()
    while True:
        response = input("\nWould you like to find the ROI on another property? (Enter 'yes' to continue, or 'no' to quit) ")
        
        if response.lower() == 'yes':
            findReturnOnInvestment()
            break
            
        elif response.lower() == 'no':
            print('Thank you for using our ROI calculator. Happy investing!')
            break
        
        else:
            response
            print('Type a valid command.')
            

findReturnOnInvestment()
    
        
        
```
